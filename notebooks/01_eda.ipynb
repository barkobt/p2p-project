# Telco Customer Churn — Keşifsel Veri Analizi (EDA)

Bu notebook veri setini anlamak, eksik değerleri tespit etmek ve churn dağılımını incelemek için kullanılır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. Veri Yükleme

```python
import sys
sys.path.append('..')

from src.preprocess import load_raw_data

df = load_raw_data()
print(f'Satır: {df.shape[0]}, Sütun: {df.shape[1]}')
df.head()
```

## 2. Temel İstatistikler

```python
df.describe(include='all')
```

```python
# Eksik değer analizi
missing = df.isnull().sum()
missing[missing > 0]
```

## 3. Hedef Değişken Dağılımı (Churn)

```python
churn_counts = df['Churn'].value_counts()
churn_pct    = df['Churn'].value_counts(normalize=True) * 100

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
churn_counts.plot.bar(ax=axes[0], color=['steelblue', 'salmon'])
axes[0].set_title('Churn Sayısı')
axes[0].set_xlabel('')

axes[1].pie(churn_counts, labels=['No Churn', 'Churn'], autopct='%1.1f%%',
            colors=['steelblue', 'salmon'])
axes[1].set_title('Churn Oranı')

plt.tight_layout()
plt.savefig('../reports/figures/churn_distribution.png', bbox_inches='tight')
plt.show()
```

## 4. Sayısal Değişken Dağılımları

```python
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col in zip(axes, numeric_cols):
    sns.histplot(data=df, x=col, hue='Churn', kde=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.savefig('../reports/figures/numeric_distributions.png', bbox_inches='tight')
plt.show()
```

## 5. Kategorik Değişkenler

```python
cat_cols = ['Contract', 'InternetService', 'PaymentMethod', 'gender']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
for ax, col in zip(axes.flatten(), cat_cols):
    churn_rate = df.groupby(col)['Churn'].mean().sort_values(ascending=False)
    churn_rate.plot.bar(ax=ax)
    ax.set_title(f'{col} — Churn Oranı')
    ax.set_ylabel('Churn Oranı')
    ax.set_xlabel('')
plt.tight_layout()
plt.savefig('../reports/figures/categorical_churn_rates.png', bbox_inches='tight')
plt.show()
```

## 6. Korelasyon Matrisi (Sayısal)

```python
corr = df[numeric_cols + ['Churn']].corr()
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=ax)
ax.set_title('Korelasyon Matrisi')
plt.tight_layout()
plt.savefig('../reports/figures/correlation_matrix.png', bbox_inches='tight')
plt.show()
```
